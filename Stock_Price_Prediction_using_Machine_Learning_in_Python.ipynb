```python
# Check if 'Close' and 'Adj Close' columns are identical
df[df['Close'] == df['Adj Close']].shape
```

```python
# Drop the 'Adj Close' column as it is redundant
df = df.drop(['Adj Close'], axis=1)
```

```python
# Check for missing values in the dataset
df.isnull().sum()
```

```python
# Define features for analysis
features = ['Open', 'High', 'Low', 'Close', 'Volume']
```

```python
# Plot distribution of features
plt.subplots(figsize=(20,10))
for i, col in enumerate(features):
  plt.subplot(2,3,i+1)
  sb.distplot(df[col])  # Distribution plot for each feature
plt.show()
```

```python
# Plot boxplots for features to identify outliers
plt.subplots(figsize=(20,10))
for i, col in enumerate(features):
  plt.subplot(2,3,i+1)
  sb.boxplot(df[col])  # Boxplot for each feature
plt.show()
```

```python
# Split the 'Date' column into 'day', 'month', and 'year'
splitted = df['Date'].str.split('/', expand=True)
df['day'] = splitted[1].astype('int')
df['month'] = splitted[0].astype('int')
df['year'] = splitted[2].astype('int')
df.head()  # Display the updated dataframe
```

```python
# Create a new column 'is_quarter_end' to indicate if the month is a quarter-end
df['is_quarter_end'] = np.where(df['month']%3==0,1,0)
df.head()
```

```python
# Group data by year and calculate mean for each year
data_grouped = df.drop('Date', axis=1).groupby('year').mean()
```

```python
# Plot mean values of 'Open', 'High', 'Low', and 'Close' prices by year
plt.subplots(figsize=(20,10))
for i, col in enumerate(['Open', 'High', 'Low', 'Close']):
  plt.subplot(2,2,i+1)
  data_grouped[col].plot.bar()  # Bar plot for each feature
plt.show()
```

```python
# Compare mean values for quarter-end vs non-quarter-end
df.drop('Date', axis=1).groupby('is_quarter_end').mean()
```

```python
# Create new features: 'open-close' and 'low-high'
df['open-close'] = df['Open'] - df['Close']
df['low-high'] = df['Low'] - df['High']
```

```python
# Create target variable: 1 if next day's closing price is higher, else 0
df['target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

```python
# Plot a heatmap to visualize highly correlated features
plt.figure(figsize=(10, 10))
sb.heatmap(df.drop('Date', axis=1).corr() > 0.9, annot=True, cbar=False)
plt.show()
```

```python
# Define features and target variable for modeling
features = df[['open-close', 'low-high', 'is_quarter_end']]
target = df['target']
```

```python
# Standardize the features
scaler = StandardScaler()
features = scaler.fit_transform(features)
```

```python
# Split the data into training and validation sets
X_train, X_valid, Y_train, Y_valid = train_test_split(
    features, target, test_size=0.1, random_state=2022)
print(X_train.shape, X_valid.shape)  # Output: Shapes of training and validation sets
```

```python
# Define models to be trained
models = [LogisticRegression(), SVC(kernel='poly', probability=True), XGBClassifier()]
```

```python
# Train and evaluate models
for i in range(3):
  models[i].fit(X_train, Y_train)  # Train the model
  print(f'{models[i]} : ')
  print('Training Accuracy : ', metrics.roc_auc_score(
    Y_train, models[i].predict_proba(X_train)[:,1]))  # Training ROC-AUC score
  print('Validation Accuracy : ', metrics.roc_auc_score(
    Y_valid, models[i].predict_proba(X_valid)[:,1]))  # Validation ROC-AUC score
  print()
```

```python
# Plot confusion matrix for the first model (Logistic Regression)
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_estimator(models[0], X_valid, Y_valid)
plt.show()
```
